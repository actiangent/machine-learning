3. Taksonomi Machine Learning terbagi menjadi 3 kelompok utama:
  1. Supervised Learning  
  Supervised learning adalah kategori di mana algoritma machine learning dilatih menggunakan data yang sudah diberi label yang berarti setiap input data disertai dengan output yang diinginkan. Tujuan dari supervised learning adalah untuk mempelajari hubungan atau fungsi yang memetakan input ke output berdasarkan contoh yang diberikan.  
  Contoh:
    * Diagnostik Medis: Memprediksi apakah seorang pasien memiliki penyakit tertentu berdasarkan data medis mereka.
  2. Unsupervised Learning
  Unsupervised learning adalah kategori di mana algoritma dilatih menggunakan data yang tidak diberi label. Algoritma mencoba mempelajair data tanpa panduan dari output yang diinginkan.  
  Contoh:
    * Segmentasi Pelanggan: Mengelompokkan pelanggan berdasarkan pola pembelian mereka untuk strategi pemasaran yang lebih efektif.
  3. Reinforcement Learning
  Reinforcement learning adalah metode di mana algoritma belajar melalui interaksi dengan lingkungan. Algoritma mendapatkan umpan balik dalam bentuk reward atau penalti berdasarkan tindakan yang diambil, dan tujuannya adalah untuk memaksimalkan total reward sepanjang waktu.  
  Contoh:
    * Robotika: Mengajarkan robot untuk berjalan atau melakukan tugas tertentu melalui trial and error.

  c) Hari apa biaya lebih dari 110000?  
  d) Siapa yang paling banyak datang ke kampus?

```python
dataset_student = {
    "Hari": ["Senin", "Selasa", "Rabu", "Kamis", "Jumat", "Sabtu", "Minggu"],
    "Datang": [2, 3, 4, 1, 2, 5 , 2],
    "Biaya": [30_000, 35_000, 25_000, 15_000, 20_000, 30_000, 35_000],
    "Mahasiswa": ["Ani", "Budi", "Jono", "Lono", "Joni", "Ani", "Budi"]
}

df = pd.DataFrame(dataset_student)
df
```

```python
df["Biaya"] = df["Biaya"] * df["Datang"]
df
```

```python
# Berapa rata-rata mahasiswa datang pada minggu ini?
mean_students_came_week = df["Datang"].mean()
print("Berapa rata-rata mahasiswa datang pada minggu ini?", str(mean_students_came_week), sep=" ")
floor_mean_students_came_week = math.floor(mean_students_came_week) # math.floor() untuk menghindari koma, pembulatan ke bawah
print("Berapa pembulatan rata-rata mahasiswa datang pada minggu ini?", str(floor_mean_students_came_week), sep=" ")
print("\n")

# Kapan biaya tertinggi terjadi?
max_cost_day = df.loc[df["Biaya"].idxmax(), "Hari"]
print("Kapan biaya tertinggi terjadi?", str(max_cost_day), sep=" ")
print("\n")

# Hari apa biaya lebih dari 110000?
day_more_than_110k = df.loc[df.index[df['Biaya'] > 110_000][0], "Hari"]
print("Hari apa biaya lebih dari 110000?", str(day_more_than_110k), sep=" ")
print("\n")

# Siapa yang paling banyak datang ke kampus?
comes_the_most = df.loc[df["Datang"].idxmax(), "Mahasiswa"]
print("Siapa yang paling banyak datang ke kampus?", str(comes_the_most), sep=" ")
print("\n")

# Siapa yang datang pada hari minggu?
comes_on_sunday = df.loc[df.index[df['Hari'] == "Minggu"][0], "Mahasiswa"]
print("Siapa yang datang pada hari minggu?", str(comes_on_sunday), sep=" ")
print("\n")

# Berapa biaya tertinggi dan terendah?
max_cost = df["Biaya"].max()
min_cost = df["Biaya"].min()
print("Berapa biaya tertinggi dan terendah?", "Biaya Tertinggi:", str(max_cost), "dan", "Biaya Terendah:", str(min_cost), sep=" ")
print("\n")

# Berapa frekuensi datang tertinggi dan terendah?
comes_most = df["Datang"].max()
comes_least = df["Datang"].min()
print("Berapa frekuensi datang tertinggi dan terendah?", "Datang Tertinggi:", str(comes_most), "kali", "dan", "Datang Terendah:", str(comes_least), "kali", sep=" ")
```
